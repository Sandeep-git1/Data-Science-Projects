Infants have more bones than adults, but as the baby develops, groups of bones fuse together. This is detailed in the 'fused_from' column of our data.

```python
import matplotlib.pyplot as plt
```

```python
# Count of bones by region
df_human['region'].value_counts().plot(kind='bar', figsize=(8,5), title="Bone Counts by Region")
plt.show()
```

```python
# Count of bones by subregion
df_human['subregion'].value_counts().plot(kind='barh', figsize=(8,5), title="Bone Counts by Subregion")
plt.show()
```

```python
# Pie chart of side
df_human['side'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title="Bones by Side")
plt.ylabel('')
plt.show()
```

```python
# Scatter plot: fused_from vs index
df_human.reset_index().plot.scatter(x='index', y='fused_from', figsize=(8,5), title="Fused_from across Bones")
plt.show()
```
